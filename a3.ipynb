```python
def grade(total):
    if(total>200):
        return "A+"
    elif(total>180):
        return "A"
    elif(total>170):
        return "B"
    elif(total>150):
        return "C"
    else:
        return "F"
```

```python
df["Grade"]=df["Total"].apply(grade)
```
